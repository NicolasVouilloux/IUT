## Exercice 1

Créer une fonction methode_dichotomie(f,a,b,epsilon) qui retourne une
approximation d’un nombre x ∈ [a, b] tel que f (x) = 0 à ε près grâce à la méthode de
dichotomie.
Cette fonction s’appliquera à des fonctions f et des points a et b tels que f (a) < 0 et
f (b) > 0.
On inclura une condition d’arrêt pour retourner une valeur dès que |yn − xn| < ε.

```python
import math
```

```python
def f0(x):
    return (x-2)
```

```python
def methode_dichotomie(f,a:float,b:float,epsilon:float):
    x=a
    y=b
    while (abs(y-x)>=epsilon):
        if (f((x+y)/2)<=0):
            x=(x+y)/2
        else :
            y=(x+y)/2
    return (x+y)/2
```

```python
x=methode_dichotomie(f0,1.0,10.0,10e-5)
print(x)
```
